<h2 id="plot">Representando datos</h2>

To visualize the temperature data we have, we’ll first create a simple plot of
the daily highs using matplotlib, as shown here: Tenmso que definir primero la variable highs, que es una lista de temeraturas máximas

```python
from matplotlib import pyplot as plt
# Almacenamos las altas temperaturas en highs.
```

```python
fig = plt.figure(dpi=64, figsize=(10, 6))
plt.plot(highs, c='red')
# Format plot.
plt.title("Altas temperaturas en año 2018", fontsize=24)
plt.xlabel('', fontsize=16)
plt.ylabel("Temperature (F)", fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=16)
plt.show()
```

Deberíamos visualizar la temperatura en grados centígrados: pasamos definir una nueva variable highsC que continene la lista de temperaturas en celsius.

```python
# definimos temperaturas en grados

fig = plt.figure(dpi=64, figsize=(10, 6))
plt.plot(highsC, c='red')
# Format plot.
plt.title("Altas temperaturas en año 2018", fontsize=24)
plt.xlabel('', fontsize=16)
plt.ylabel("Temperatura en grados Cº", fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=16)
plt.show()
```

Let’s add dates to our graph to make it more useful. The first date from the
weather data file is in the second row of the file. 
we need a way to convert the
string '2014-7-1' to an object representing this date. We can construct an
object representing July 1, 2014, using the strptime() method from the
datetime module. Let’s see how strptime() works

```python
from datetime import datetime
first_date = datetime.strptime('2014-7-1', '%Y-%m-%d')
print(first_date)
```

The <code> strptime()</code> method can take a variety of arguments to determine
how to interpret the date <br/>
<img src="src/date_table.png" width="500"></img>

Knowing how to process the dates in our CSV file, we can now improve our
plot of the temperature data by extracting dates for the daily highs and
passing the dates and the highs to plot(), as shown here:

```python
# Transformamos la lista de strings en una lista de dates

liststringdate = df['DATE']
listdate = [datetime.strptime(item, "%Y-%m-%d") for item in liststringdate]
#print(liststringdate)
```

```python
fig = plt.figure(dpi=64, figsize=(10, 6))
plt.plot(listdate, highsC, c='red')

# Format plot.
plt.title("Altas temperaturas en año 2018", fontsize=24)
plt.xlabel('', fontsize=16)
fig.autofmt_xdate() #draws the date labels diagonally to prevent them from overlapping.
plt.ylabel("Temperatura en grados Cº", fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=16)
plt.show()
```
